```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import stats
```

```python
file_path = "life_expectancy.csv"

life_df = pd.read_csv(file_path)
```

```python
#Only grab relevant columns
life_df1 = life_df[['State','Life Expectancy']]

#drop rows with blank values
life_df1 = life_df1.dropna()

life_df1
```

```python
#take average of Life Expectancy in 2010-2015 by State
life_df2 = round(life_df1.groupby('State').mean(),2)
life_df2 = life_df2.reset_index()

life_df2.head()
```

```python
#import locations csv
file = "locations1.csv"

location_df = pd.read_csv(file)
location_df = location_df.dropna()
location_df = location_df.astype({'Location Id': 'int64'})
loc_renamed_df = location_df.rename(columns = {'US states and territories A-Z':'State'})
loc_renamed_df.head()
```

```python
#left join by location_df State and drop Columns that aren't needed

merged_left = pd.merge(left=loc_renamed_df, right=life_df2, how='left', left_on='State', right_on='State')
merged_left.head()
```

```python
#drop State Column
merged_left1 = merged_left.drop(['State'], axis =1)
merged_left1.head()
```
